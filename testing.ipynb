```python
from main import DynamicMa as dm
import yfinance as yf
from matplotlib import pyplot as plt
```

```python
spx = yf.download("^SPX")["Close"]
```

```python
strat = dm(spx, range(10,201,10))
```

```python
all_positions = strat.get_all_postions()
net_positions = strat.get_net_position(all_positions)
strat_daily_return = strat.get_daily_return(net_positions)
strat_cumulative_return = strat.get_cum_return(strat_daily_return)
buy_and_hold_daily_return = spx.pct_change()
buy_and_hold_cumulative_return = (buy_and_hold_daily_return + 1).cumprod()
```

```python
fig, ax = plt.subplots(figsize= (20,10))
ax.plot(strat_cumulative_return, color = "red", label = "strat")
ax.plot(buy_and_hold_cumulative_return, color = "blue", label = "buy and hold")
ax.set_yscale("log")
ax.legend()
```
